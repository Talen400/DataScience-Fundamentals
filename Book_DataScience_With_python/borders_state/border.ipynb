```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
geojson_file = 'gadm41_BRA_1.json'

gdf = gpd.read_file(geojson_file)

gdf
```

```python
geokmz_file = 'gadm41_BRA_0.kml'

gdf2 = gpd.read_file(geokmz_file)

gdf2
```

```python
gdf.plot()
gdf2.plot()
```

```python

# Criar uma figura e um eixo
fig, ax = plt.subplots(figsize=(12, 8))

# Plotar os países


# Plotar as cidades
gdf2.plot(ax=ax, color='red', markersize=5, alpha=0.7, edgecolor='black')

gdf.plot(ax=ax, facecolor='none', linewidth=1.5)

# Adicionar títulos e labels
ax.set_title('Countries and Cities')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Ajustar o layout
plt.tight_layout()

# Mostrar o mapa
plt.show()
```
